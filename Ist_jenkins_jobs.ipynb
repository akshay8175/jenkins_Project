# My first jenkins Job

### Step 1. Crete jenkins job 



### Step 2. Source Code Management
- Choose Git.
- Add your repository URL.
- (Optional) Add credentials if it's a private repo.
- select Branch name according to git repo
  


### Step 3 

- Build Environment : Add anything specific like cleaning workspace before build, etc.

- Build Steps : Click Add build step > Execute shell (Linux/macOS) or Execute Windows batch command.
- Add a simple command:
> echo "Hello from Jenkins!"



### Step 4: Save and Build

- Click Save.
- Then click `Build Now` on the project page.
